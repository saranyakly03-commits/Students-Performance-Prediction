```python
#Loading and exploring dataset
```

```python
import pandas as pd
data=pd.read_csv("student_performance_dataset.csv")
data.head()
```

```python
data.info()
```

```python
#Data Cleaning
```

```python
#Check null values
data.isnull().sum()
```

```python
#Convert Categorical data into Numerical data
from sklearn.preprocessing import LabelEncoder

binary_columns=["Gender","Internet_Access_at_Home","Extracurricular_Activities","Pass_Fail"]
le=LabelEncoder()
for col in binary_columns:
    data[col]=le.fit_transform(data[col])
data=pd.get_dummies(data,columns=["Parental_Education_Level"],drop_first=True)

print("\nCleaned Data")
data.head()
                    
```

```python
#Performing Exploratory Data Analysis
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
sns.histplot(data["Final_Exam_Score"],bins=15,color="blue")
plt.title("Distribution of Final Exam Scores")
plt.xlabel("Final Exam Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Study_Hours_per_Week",y="Final_Exam_Score",data=data,color="green")
plt.title("Scatter Plots of Study Hours Per Week and Final Exam Scores")
plt.xlabel("Study Hours Per Week")
plt.ylabel("Final Exam Score")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Parental_Education_Level_High School",y="Final_Exam_Score",data=data,color="orange")
plt.title("Box Plot of Final Exam Scores by Parental Education(High School)")
plt.xlabel("Final Exam Score")
plt.ylabel("Parental Education Level")
plt.show() 


plt.figure(figsize=(8,5))
sns.boxplot(x="Parental_Education_Level_Masters",y="Final_Exam_Score",data=data,color="orange")
plt.title("Box Plot of Final Exam Scores by Parental Education(Masters)")
plt.xlabel("Final Exam Score")
plt.ylabel("Parental Education Level")
plt.show() 


plt.figure(figsize=(8,5))
sns.boxplot(x="Parental_Education_Level_PhD",y="Final_Exam_Score",data=data,color="orange")
plt.title("Box Plot of Final Exam Scores by Parental Education(PhD)")
plt.xlabel("Final Exam Score")
plt.ylabel("Parental Education Level")
plt.show() 

```

```python
#FEATURE ENGINEERING AND CORRELATION
```

```python
import numpy as np

data["Study_Efficiency_Score"]=data["Final_Exam_Score"]/data["Study_Hours_per_Week"]

print("\nData with Study Efficiency Score")
print(data[["Final_Exam_Score","Study_Hours_per_Week","Study_Efficiency_Score"]].head())
```

```python
plt.figure(figsize=(10,8))
corr_matrix=data.corr(numeric_only=True)
sns.heatmap(corr_matrix,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

```python
#Predictive Modeling 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.preprocessing import StandardScaler
```

```python
x=data.drop(columns=["Student_ID","Final_Exam_Score"])
y=data["Final_Exam_Score"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

Linear Regression

```python
linear_model=LinearRegression()
linear_model.fit(x_train_scaled,y_train)
linear_preds=linear_model.predict(x_test_scaled)
```

```python
print("Linear Regression Performance\n")
print("Mean Absolute Error:",(mean_absolute_error(y_test,linear_preds)))
print("Mean Squared Error:",(mean_squared_error(y_test,linear_preds)))
print("R2 Score:",r2_score(y_test,linear_preds))
```

KNN

```python
knn_model=KNeighborsRegressor(n_neighbors=5)
knn_model.fit(x_train_scaled,y_train)
knn_preds=knn_model.predict(x_test_scaled)
```

```python
print("\nKNN Performance\n")
print("Mean Absolute Error:",(mean_absolute_error(y_test,knn_preds)))
print("Mean Squared Error:",(mean_squared_error(y_test,knn_preds)))
print("R2 Score:",r2_score(y_test,knn_preds))
```

Decision Tree

```python
dt_model = DecisionTreeRegressor(
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42
)

dt_model.fit(x_train_scaled, y_train)

dt_preds = dt_model.predict(x_test_scaled)
```

```python
print("\nDecision Tree Performance\n")
print("Mean Absolute Error:",(mean_absolute_error(y_test,dt_preds)))
print("Mean Squared Error:",(mean_squared_error(y_test,dt_preds)))
print("R2 Score:",r2_score(y_test,dt_preds))
```

```python
rf_model = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42
)

rf_model.fit(x_train_scaled, y_train)

rf_preds = rf_model.predict(x_test_scaled)
```

```python
print("\nRandom Forest Performance\n")
print("Mean Absolute Error:",(mean_absolute_error(y_test,rf_preds)))
print("Mean Squared Error:",(mean_squared_error(y_test,rf_preds)))
print("R2 Score:",r2_score(y_test,rf_preds))
```

Support Vector Regression(SVR)

```python
svr=SVR(kernel="rbf")
svr.fit(x_train_scaled,y_train)

svr_preds=svr.predict(x_test_scaled)
```

```python
print("\nSVR\n")
print("Mean Absolute Error:",(mean_absolute_error(y_test,svr_preds)))
print("Mean Squared Error:",(mean_squared_error(y_test,svr_preds)))
print("R2 Score:",r2_score(y_test,svr_preds))
```

Gradient Boosting Regression

```python
gbr=GradientBoostingRegressor(random_state=42)
gbr.fit(x_train,y_train)
gbr_preds=gbr.predict(x_test)
```

```python
print("\nGradient Boosting Regression\n")
print("Mean Absolute Error:",(mean_absolute_error(y_test,gbr_preds)))
print("Mean Squared Error:",(mean_squared_error(y_test,gbr_preds)))
print("R2 Score:",r2_score(y_test,gbr_preds))
```

```python
#find best algorithm
results = {
    "Linear Regression":r2_score(y_test,linear_preds),
    "KNN":r2_score(y_test,knn_preds),
    "Decision Tree": r2_score(y_test, dt_preds),
    "Random Forest": r2_score(y_test, rf_preds),
    "SVR":r2_score(y_test,svr_preds),
    "Gradient Boosting Regression":r2_score(y_test,gbr_preds)
}

best_algorithm = max(results, key=results.get)
best_algorithm
```
